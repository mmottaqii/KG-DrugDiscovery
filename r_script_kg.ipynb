```r
# Set the directory where the files are stored
file_dir <- '~/kg/datasets'

# List all .rds files in the directory
files <- list.files(path = file_dir, pattern = "\\.rds$", full.names = TRUE)

# Initialize a list to hold the names of the data objects
data_object_names <- list()

# Loop over each file, read it, and assign to a variable named based on the file name
for (file_path in files) {
  # Extract the base name of the file (without directory and extension)
  file_name <- basename(file_path)
  # Extract the part of the name before the underscore
  obj_name <- unlist(strsplit(file_name, "_"))[1]
  # Read the RDS file
  assign(obj_name, readRDS(file_path))
  # Add the object name to the list
  data_object_names[[length(data_object_names) + 1]] <- obj_name
}

# Print the names of all loaded datasets
print(data_object_names)
```

```r
# Assuming 'data_object_names' holds the names of the loaded datasets
for (dataset_name in data_object_names) {
  # Access the dataset dynamically using get()
  dataset <- get(dataset_name)
  
  # Print the number of drugs
  cat(sprintf("Number of drugs in %s: %d\n", dataset_name, length(drugNames(dataset))))
  
  # Print the structure of the sensitivity slot
  cat(sprintf("Structure of the sensitivity slot in %s:\n", dataset_name))
  print(str(sensitivitySlot(dataset)))
  
  # Print the head of the sensitivity profiles
  cat(sprintf("Head of the sensitivity profiles in %s:\n", dataset_name))
  print(head(sensitivityProfiles(dataset)))
  
  # Summarize sensitivity profiles for 'ic50_recomputed'
  sensitivity_profiles <- summarizeSensitivityProfiles(dataset, 'ic50_recomputed')
  
  # Compute the number of non-NA values (valid entries) for each drug
  non_na_counts <- apply(sensitivity_profiles, 1, function(x) sum(!is.na(x)))
  
  # Calculate the average number of cell lines per drug with non-NA values
  average_non_na <- mean(non_na_counts)
  
  # Print the average number
  cat(sprintf("Average number of non-NA cell lines per drug in %s: %f\n", dataset_name, average_non_na))
  cat("--------------------------------------------------\n")
}
```

```r
# Summarize AAC recomputed sensitivity profiles
# ic50_published	aac_published	amax_published	aac_recomputed	ic50_recomputed	HS	E_inf	EC50
sensitivity_profiles = summarizeSensitivityProfiles(DGSC_2, sensitivity.measure='aac_recomputed')
```

```r
sensitivity_profiles
```

```r
# Remove columns that contain only NA values
sensitivity_profiles_clean = sensitivity_profiles[, apply(sensitivity_profiles, 2, function(x) !all(is.na(x)))]

# Convert AAC to AUC by subtracting from 1
sensitivity_profiles_auc = 1 - sensitivity_profiles_clean
```

```r
sensitivity_profiles_auc
```

```r
# check the normality for drugs
```

```r
# Assuming 'sensitivity_profiles_auc' has your AUC data with drugs as row names

# Correct application of the Shapiro-Wilk test with accurate drug naming
results <- apply(sensitivity_profiles_auc, 1, function(drug_values) {
  # Remove NA values for the normality test
  valid_values <- na.omit(drug_values)

  # Ensure there are enough values to perform the test
  if (length(valid_values) > 3) {
    test <- shapiro.test(valid_values)
    p_value <- test$p.value
  } else {
    p_value <- NA  # Not enough data to compute the test
  }

  return(p_value)
})

# Create a named vector with drug names as names of the vector
names(results) <- rownames(sensitivity_profiles_auc)

# Output the results for each drug
results_df <- data.frame(Drug = names(results), P_Value = results, row.names = NULL)
print(results_df)

# Determine how many drugs have p-values greater than 0.05 and 0.99
normal_05 <- sum(!is.na(results) & results > 0.05, na.rm = TRUE)
normal_99 <- sum(!is.na(results) & results > 0.99, na.rm = TRUE)

# Output the count of normally distributed drugs at different thresholds
cat("Number of drugs with normal distribution at p > 0.05:", normal_05, "\n")
cat("Number of drugs with normal distribution at p > 0.99:", normal_99, "\n")
```

```r
# Assuming 'sensitivity_profiles_auc' has your AUC data with drugs as row names
# and it is already loaded into your R environment

# Open a PNG file to save the plots
png("DGSC_v2_Drugs_Distribution.png", width=5500, height=5200, res=900)  # Set resolution to 900 DPI

# Create a plot layout
par(mfrow=c(2,2))  # Set up the plotting area to have 2 rows and 2 columns

# Loop through each of the first four drugs and plot
for (i in 1:4) {
  drug_data <- na.omit(sensitivity_profiles_auc[i, ])  # Remove NA values for plotting

  # Create a histogram with a density plot overlaid
  hist(drug_data, breaks=30, main=paste("Histogram for", rownames(sensitivity_profiles_auc)[i]), xlab="AUC Values", probability = TRUE, col=rgb(0, 0, 1, 0.5))  # Light blue histogram
  lines(density(drug_data), col="red", lwd=2)  # Red density line
}

# Reset the plotting parameters
par(mfrow=c(1,1))

# Close the PNG file
dev.off()
```

```r
# Assuming 'sensitivity_profiles_auc' has your AUC data with drugs as row names

# Step 1: Apply Log3 Transformation
# Add a small constant to avoid log of zero if zero values are present
sensitivity_profiles_log3 <- log(sensitivity_profiles_auc + 0.001) / log(2)

# Step 2: Check for Normality on Transformed Data
normality_results_log3 <- apply(sensitivity_profiles_log3, 1, function(x) {
  x <- na.omit(x)  # Remove NA values
  if (length(x) > 3) {  # Ensure enough data points for the test
    test <- shapiro.test(x)
    return(test$p.value)
  } else {
    return(NA)  # Not enough data to perform the test
  }
})

# Print results
results_log3 <- data.frame(Drug = rownames(sensitivity_profiles_auc), P_Value = normality_results_log3, row.names = NULL)
print(results_log3)

# Determine how many drugs have p-values indicating normality
normal_05_log3 <- sum(!is.na(normality_results_log3) & normality_results_log3 > 0.05, na.rm = TRUE)
normal_99_log3 <- sum(!is.na(normality_results_log3) & normality_results_log3 > 0.99, na.rm = TRUE)

# Print the number of drugs with normally distributed values at different thresholds
cat("Number of drugs with normal distribution at p > 0.05 after log3 transformation:", normal_05_log3, "\n")
cat("Number of drugs with normal distribution at p > 0.99 after log3 transformation:", normal_99_log3, "\n")
```

```r
# Z-score
# is the dataset normal?
```

```r
# Compute z-scores for each drug, ignoring NA values
sensitivity_profiles_z = apply(sensitivity_profiles_auc, 1, function(x) {
  # Calculate mean and standard deviation ignoring NA values
  drug_mean = mean(x, na.rm = TRUE)
  drug_sd = sd(x, na.rm = TRUE)

  # Return z-score if sd is not zero; otherwise return NA to handle constant rows
  if (drug_sd > 0) {
    return((x - drug_mean) / drug_sd)
  } else {
    return(rep(NA, length(x)))
  }
})

# Convert the result back into a data frame, with drugs as row names
sensitivity_profiles_z = as.data.frame(t(sensitivity_profiles_z))
rownames(sensitivity_profiles_z) = rownames(sensitivity_profiles_auc)
```
